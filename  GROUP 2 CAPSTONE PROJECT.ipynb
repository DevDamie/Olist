```python
##Importing our Libraries


import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
import numpy as np
from datetime import datetime
from dateutil import parser


%matplotlib inline
```

```python
#Calling our tables

order_table = pd.read_csv("olist_orders_dataset.csv")
customer_table = pd.read_csv("olist_customers_dataset.csv")
order_items_table = pd.read_csv("olist_order_items_dataset.csv")
order_payments_table = pd.read_csv("olist_order_payments_dataset.csv")
product_table = pd.read_csv("olist_products_dataset_cleaned.csv")
order_reviews_table = pd.read_csv("olist_order_reviews_dataset.csv")
sellers_table = pd.read_csv("olist_sellers_dataset.csv")
```

```python
order_table.head(20)
```

```python
#changing the data type of some columns to datetime


order_table["order_purchase_timestamp"] = pd.to_datetime(order_table["order_purchase_timestamp"])
order_table["order_approved_at"] = pd.to_datetime(order_table["order_approved_at"])
order_table["order_delivered_carrier_date"] = pd.to_datetime(order_table["order_delivered_carrier_date"])
order_table["order_delivered_customer_date"] = pd.to_datetime(order_table["order_delivered_customer_date"])
order_table["order_estimated_delivery_date"] = pd.to_datetime(order_table["order_estimated_delivery_date"])
```

```python
order_table.head(5)
```

```python
#dropping 2016 data

ORDER_TABLE=order_table.loc[(order_table['order_purchase_timestamp'] >='01-01-2017')
                                    & (order_table['order_purchase_timestamp'] < '01/01/2019')]
ORDER_TABLE
```

```python
#To know the first order date in the year 2017, and the last order date in 2018

max_time = ORDER_TABLE.order_purchase_timestamp.max()
min_time = ORDER_TABLE.order_purchase_timestamp.min()
number_records = str(len(ORDER_TABLE))
num_unique_orders = str(ORDER_TABLE['order_id'].nunique())
num_unique_customers = str(ORDER_TABLE['customer_id'].nunique())

print(min_time)
print(max_time)
print(number_records)
print(num_unique_orders)
print(num_unique_customers)
```

```python
#Customer_table

customer_table.head(10)
```

```python
## Joining our orders table with customer table  # type of join = Left join = ORDER CUSTOMER TABLE
## WITH Order reviews and Order payement table; the four tables is now called 'ALL_ORDER_TABLE'

order_customer_table = ORDER_TABLE.merge(customer_table, how='left', on="customer_id")
order_customer_review_table = order_customer_table.merge(order_reviews_table, how='left', on="order_id")
all_order_table = order_customer_review_table.merge(order_payments_table, how='left', on="order_id")

all_order_table.head(10)
```

```python
## Checking our Null values

Null_values = pd.isnull(all_order_table).sum()
Null_values
```

```python
## Removing our Null Values

all_order_table = all_order_table.dropna(how= 'any', axis=0)
all_order_table
```

```python
# aggregating payment_value to get the order total and merging it back onto the main dataframe

order_total =  all_order_table.groupby('order_id').agg({'payment_value':'sum'}).reset_index()
order_total.columns = ['order_id', 'total_order_value']
all_order_table = all_order_table.merge(order_total, how='left', on="order_id")
all_order_table.head(10)
```

```python
# converting day object to a timestamp so we can truncate to the day 
# (rather than having all the trailing hours + minutes)

all_order_table['order_purchase_timestamp_day'] = pd.to_datetime(all_order_table['order_purchase_timestamp'],\
                                                                format='%Y-%m-%d %H:%M:%S')
all_order_table['order_purchase_timestamp_month'] = all_order_table['order_purchase_timestamp_day'].dt.strftime('%b %Y')

# strip dataframe to contain just order month and payment value (e.g. only fields we want to aggregate)

revenue = all_order_table[['order_purchase_timestamp_month', 'payment_value', 'order_id']]

# now we're going to aggregate payment_value and order_id by the month to get the montly revenue and 
# monthly number of orders

revenue =  revenue.groupby('order_purchase_timestamp_month').\
    agg({'payment_value':'sum', 'order_id': pd.Series.nunique}).reset_index()
revenue.head(20)
```

Getting customer order numbers
how many repeat customers we have between 2017 and 2018


```python
customer_sales_summary = all_order_table.groupby('customer_unique_id').\
    agg({'order_purchase_timestamp_month':'min', 'payment_value': 'sum', 'order_id':  pd.Series.nunique}).reset_index()
customer_sales_summary.columns = ['customer_unique_id', 'first_purchase_month', 'total_spend', 'num_orders']
customer_sales_summary.head(50)
```

```python
# making a pivot to see how many customers fall within each order number bucket
# e.g. 93099 customers have placed 1 order

num_cus_by_num_orders = customer_sales_summary.groupby('num_orders')['customer_unique_id'].count().reset_index()
num_cus_by_num_orders.head(15)
```

```python
# created pivot to get the total spend per "num order" grouping
# question we are  trying to answer, what is the median order price for customers who have X number of orders
# e.g. Customers who have 2 orders have an average, median order price of $114.30 (228.59/2) 

med_spend_by_num_orders = customer_sales_summary.groupby('num_orders').\
    agg({'total_spend':'median',  'customer_unique_id': 'count'}).reset_index()
med_spend_by_num_orders.columns = ['num_orders', 'median_spend', 'num_customers']
med_spend_by_num_orders.head()
```

```python
# Trying to labeling data with easy to read legends for our charts

customer_sales_summary['num_order_category'] = ''
customer_sales_summary.loc[customer_sales_summary['num_orders'] == 1, 'num_order_category'] = '1 order'
customer_sales_summary.loc[customer_sales_summary['num_orders'] == 2, 'num_order_category'] = '2 orders'
customer_sales_summary.loc[customer_sales_summary['num_orders'] == 3, 'num_order_category'] = '3 orders'
customer_sales_summary.loc[customer_sales_summary['num_orders'] > 3, 'num_order_category'] = '>3 order'

# creating the average spend per order metric 
#  average spend per order = total spend across all orders / number of orders
customer_sales_summary['average_spend_per_order'] = \
    customer_sales_summary['total_spend']/customer_sales_summary['num_orders']
```

```python
# truncating dataframe to only what we're going to use

df = customer_sales_summary[['average_spend_per_order', 'num_order_category']]


#sb.set(rc={'figure.figsize':(10,5)})
#sb.despine()

# creating box plot
bplot = sb.boxplot(y='average_spend_per_order', x='num_order_category', 
                 data=df, 
                 width=0.5,
                 showfliers=False,
                palette="Blues")
# setting chart title, axis names, and changing styling
bplot.axes.set_title("Distribution of spend by customer type",fontsize=16)
bplot.set_xlabel("Customers split by number of orders", fontsize=16)
bplot.set_ylabel("Average spend per order",fontsize=16)
bplot.tick_params(labelsize=15)
sb.set_style("ticks")
```

```python
# merging customer_sales table + all_order_table  (already translated in english) type of join = left = all_order_with_customer_history
# and dropping unwanted  columns amd also removing duplicates to keep it bit more concise
all_order_with_customer_history = all_order_table.merge(customer_sales_summary, how='left', on="customer_unique_id")
all_order_with_customer_history = all_order_with_customer_history.drop(['order_purchase_timestamp',\
    'order_approved_at', 'order_delivered_carrier_date','order_delivered_customer_date', \
    'order_estimated_delivery_date', 'customer_zip_code_prefix', 'customer_city',\
    'customer_state', 'review_id', 'review_score', 'review_comment_title','review_comment_message', \
    'review_creation_date','review_answer_timestamp', 'order_status', 'payment_sequential', 'payment_type', \
    'payment_installments', 'payment_value'],  axis=1)
all_order_with_customer_history = all_order_with_customer_history.drop_duplicates()
all_order_with_customer_history.rename(columns={'total_order_value_x':'total_order_value'}, inplace=True)
all_order_with_customer_history.head(10)
```

```python
# merging product info table with order items table, type of join = outer = product_table_category_order_table
 
product_table_category_order_table = order_items_table.merge(product_table, how='outer', on="product_id")
product_table_category_order_table = product_table_category_order_table.drop(['product_name_lenght','product_description_lenght',\
                                 'product_weight_g', 'product_length_cm','product_height_cm','product_width_cm'],  axis=1)
product_table_category_order_table.head(5)
```

```python
# merging all the tables together
all_order_info_with_item_customer_info = all_order_with_customer_history.merge(product_table_category_order_table,\
                                        how='left', on="order_id")
all_order_info_with_item_customer_info.head(10)
```

```python
# calculating item total price because shipping is broken out as a seperate price
all_order_info_with_item_customer_info['item_total_price'] = \
    all_order_info_with_item_customer_info['price'] + all_order_info_with_item_customer_info['freight_value']
```

```python
# creating a dataframe tosee how many items are in an order and what the average item price is
items_per_order = all_order_info_with_item_customer_info.groupby(['order_id', 'num_order_category', 'product_category_name']).\
    agg({'product_id':'count',  'item_total_price': 'mean'}).reset_index()
items_per_order.columns = ['order_id', 'num_order_category','num_items', 'product_category_name','avg_price']
items_per_order.head(3)
```

```python
# making a pivot to see how many customers fall within each # of items bucket
items_per_order_aggr = items_per_order.groupby(['num_items','product_category_name']).\
    agg({'avg_price':'mean',  'order_id':pd.Series.nunique}).reset_index()
items_per_order_aggr.columns = ['num_items', 'avg_price','num_orders','product_category_name']
items_per_order_aggr.head()
```

```python
items_per_order[items_per_order['num_items']==1]['avg_price'].describe()
```

```python
# creating two pivots to see top cateogries by volume and by dollar amount
top_selling_volume = all_order_info_with_item_customer_info.groupby('product_category_name')\
    ['order_id'].count().reset_index()
top_selling_volume
sort_value = top_selling_volume.sort_values(by = 'order_id', ascending = False)
sort_value.head(10)
```

```python
# creating dataframe to get the review information and dropping not relevant columns
all_order_info_with_item_customer_info_review = all_order_info_with_item_customer_info.merge(order_reviews_table, how='left', on="order_id")
all_order_info_with_item_customer_info_review = all_order_info_with_item_customer_info_review.\
    drop(['review_id', 'review_comment_title','review_comment_message','review_answer_timestamp'],  axis=1)
all_order_info_with_item_customer_info_review.head(5)
```

```python
review_by_month = all_order_info_with_item_customer_info_review.groupby(['first_purchase_month']).\
    agg({'review_score':['mean', 'median', 'count', np.std]}).reset_index()
review_by_month
```

```python
review_by_category = all_order_info_with_item_customer_info_review.groupby(['product_category_name']).\
    agg({'review_score':['mean', 'median', 'count', np.std], 'order_id': pd.Series.nunique}).reset_index()
review_by_category.head(50)
```
